### _**5. Missing Values Analysis**_

```python
missing_value= df.isnull().sum().to_frame().rename({0: 'Missing Values Number'}, axis = 1).style.background_gradient('gist_heat_r')
missing_value
```

```python
missing_val_df = ((df.isnull().sum() / df.shape[0])*100).to_frame().reset_index().rename({'index': 'columns', 0: 'Missing Values percentage'}, axis = 1)

plt.figure(figsize = (10,8))
sns.barplot(data = missing_val_df, x = 'Missing Values percentage', y = 'columns',palette='icefire')
plt.title('Missing Values percentage', fontsize = 20, fontweight = 'bold')
plt.xlabel(" ")
plt.ylabel(" ")
```

- We can clearly see that there are more than 20% of Missing data in "last_review" and "reviews_per_month" column and very few nos of missing data in 'name' and 'host_name' column

##### **Missing Values Treatments**
- 'host_name' and 'name' column has too less NAN value, we are going to impute them by forward fill.
- The NAN value of "reviews_per_month" column can be replaced by the mean value.
- The NAN value of "last_review" column can be replaced by the forward fill.
##### **What is Forwardfill?**
- Forward fill, also known as "last observation carried forward" (LOCF), is a method for filling in missing values in a time series dataset. This method involves filling in missing values with the last observed value.
- The forward fill (also known as forward propagation) method for filling in missing values involves replacing missing values with the next non-missing value in the same column. This method assumes that the values in a column are correlated with each other and that the missing value can be estimated from the surrounding values.
##### **What is mean value imputation in missing value treatment?**
- Mean value imputation is a common method for filling in missing values in a dataset. This approach involves replacing missing values with the mean value of the available data for the variable in question.

```python
df.host_name.fillna(method="ffill", inplace=True) ## Forwardfill 
df.name.fillna(method="ffill", inplace=True) ## Forwardfill 
```

```python
df["reviews_per_month"] = df["reviews_per_month"].fillna(df["reviews_per_month"].mean()) ## Mean value imputation
df.last_review.fillna(method="ffill", inplace=True) ## Forwardfill
```

### **3. Will the gained insights help creating a positive business impact?** 
- Yes it will



```python
# Computing the quartiles,iqr and upper bound to get the outliers 
q1=np.quantile(df["price"],.25)
q3=np.quantile(df["price"],.75)
iqr=q3-q1
upper_bound = q3+(1.5*iqr)
print(f"This {upper_bound} is the upper bound for outlier of price.")
```

##### **OBSERVATION**
- 334.0 is the upper bound for price anything above 334.0 in price column would be considered as a outlier.

- Below we consider outlier price data i.e data with price more than 334.0 in price column and try to see in which area's those outliers lies in and thier room types. The reason to see the different room types is because the price is dependent on room types.

```python

# filtering out the price data using above upper bound and calculating the count percent of room type and creating the seprate dataframe for it 
df_outlier_price=df[df["price"]>upper_bound]
d=df_outlier_price.groupby(["neighbourhood_group","room_type"])["id"].count().reset_index().sort_values(by="id",ascending=False)
d["Count of Room types in Percent"]=(d["id"]/d["id"].sum())*100
d.drop("id",axis=1,inplace=True) 
d.rename(columns={"neighbourhood_group":"Neighbourhood Group","room_type":"Room Type"},inplace=True)
d
```

### **1.Why did you pick the specific Table?**
- To see the highest Count of Room types in Percent according to neighborhood group and room type.

### **2. What is/are the insight(s) found from the chart?**
- From above table we can see that most of outliers are from Manhattan area and their room type is Entire home/apt which accounts to 64.1 % of the data and then followed by Brooklyn area with same room type i.e Entire home/apt with 20.8 % and so on.
